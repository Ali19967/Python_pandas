## Question 1: What percentage of all licensed sets ever released were Star Wars themed?

```python
# It is mentioned in the "Introduction" above, the 'parent_theme' column of "lego_sets" table is similar to the 'name' column of
# 'parent_themes' table. therefore we are going to use (Dataframe.merge) to combine these two dataframes.
# syntax: newdf = df1.merge(df2, how='right')

merged_df = df.merge(theme, left_on= 'parent_theme', right_on='name')
merged_df.head()

```

#### Note: In the output above, you can observe that there are two name columns, name_x and name_y. This is due to the fact that we've had name columns in both our dataframes that we have merged. To handle this, let us drop the name_y column.

```python
merged_df.drop(columns='name_y')
merged_df.head()
```

#### Again you'll see that, in output we have both the columns 'name_x' & 'name_y', this is because we have not used 'inplace=True'

```python
merged_df.drop(columns='name_y', inplace=True)
merged_df.head()

# Note: to drop 'name_y' we can also write the code as:
# merged_df = merged_df.drop(columns='name_y')
# merged_df.head()
```

#### Our next step is to filter all the records with 'is_licensed' = 'True', i.e they are licensed.

```python
licensed = merged_df[merged_df['is_licensed']]
licensed.head()
```

#### Now we need to do the percentage i.e, What percentage of themes are Star Wars?

```python
# First let us get the dataframe with parent_theme == Star Wars, to do this we again perform filtering.
Star_wars = licensed[licensed['parent_theme']== 'Star Wars']
Star_wars.head(50)
```

### Important Note: It is given to us in the instructions that 'set_num' column is critical, and a missing value indicates the set is a duplicate or invalid!
#### But the output of the code line below contains the null values as well, hence to get the correct answer we'll need to deal with the null values first.
licensed = merged_df[merged_df['is_licensed']]

```python
# Removing null values
# First let us check how many null values are present in our merged_df dataframe. 

print(merged_df['set_num'].isnull().sum())

# We can also use other ways to see number of null values, one of which is shown below using filtering and shape():
# merged_df[merged_df['set_num'].isnull()].shape
```

#### To calculate percentage we need the count of: 
##### 1) licensed records
##### 2) licensed records with parent_name == Star Wars

```python
licensed = merged_df[merged_df['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])

Star_wars = licensed[licensed['parent_theme']== 'Star Wars']

the_force = int(Star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```

### Explanation of the code in the above cell. The first line will get all the 'is_licensed' records; after that, we've dropped the rows where'set_num == NaN'. If we've simply used dropna(), then in this case, it will remove all the rows with at least one NaN value. But using dropna (subset=['set_num']) will help to remove rows where'set_num == NaN'. The third line will retrieve all records where 'parent_theme == 'Star Wars'. The 4th line stores the percentage in a variable named 'the_force'.

```python
# another way to find out the percentage.
force = int(len(Star_wars)/len(licensed)*100)
print(force)
```

## In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)? 

```python
licensed.head()
```

```python
# let us sort the licensed dataframe to reduce complexity
licensed_sorted = licensed.sort_values('year')
licensed_sorted['Count']=1   #adding a new column count with all values = 1
licensed_sorted.head(50)
```

```python
summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()
summed_df.head()
```

```python
max_df = summed_df.sort_values('Count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df
```

## From max_df above we get our answer viz Year 2017, in which Super Heroes surpassed the Star Wars

### Question 3: How many unique sets were released each year (1999-2017)

```python
# Now we are going to create a cleaned dataframe 
clean_df = merged_df[~merged_df['set_num'].isnull()]
clean_df['count'] = 1
clean_df.head()
```

```python
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year', 'count']]
sets_per_year
```

```python
# to display all records
for index, row in sets_per_year.iterrows():
    print(row['year'],row['count'])
```
